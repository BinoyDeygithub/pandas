```python
import pandas as pd
df=pd.read_csv('food delivery costs.csv')
df
df.head()


def extract(value):
    a=str(value).split(" ")
    return a[0]


df['Discounts and Offers']=df['Discounts and Offers'].apply(extract)
print(df.head())
```

```python
def extract(value):
    a=str(value).split(" ")
    return a[0]


df['Discounts and Offers']=df['Discounts and Offers'].apply(extract)
print(df.head())
```

```python
def remove(value):
    if "%" in value:
        a=value.replace('%', '')
        return float(a)
    else:
        return float(value)

df['Discounts and Offers']=df['Discounts and Offers'].apply(remove)
print(df.head())
```

```python
df.loc[(df['Discounts and Offers']<=15),'Discounts and Offers']=(df['Discounts and Offers']/100)*df['Order Value']
print(df.head())
```

```python
df['Discounts and Offers']=df['Discounts and Offers'].fillna(0)
df.head()
```

```python
df['Costs']=df['Delivery Fee']+df['Discounts and Offers']+df['Payment Processing Fee']
df.head()
```

```python
df['Profit']=df['Commission Fee']-df['Costs']
df.head()
```

```python
df['Profit'].sum()
```

```python
cost_dict=df[['Delivery Fee','Payment Processing Fee','Discounts and Offers']].sum()
cost_dict
```
